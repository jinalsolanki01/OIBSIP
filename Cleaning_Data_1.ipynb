## Data Cleaning – NYC Airbnb Open Data

### Dataset Overview and Objective

This project focuses on cleaning the NYC Airbnb Open Data dataset.
The dataset contains information about Airbnb listings in New York City, including prices, locations, room types, availability, and host details.

The objective of this task is to improve data quality by identifying and fixing common issues such as missing values, duplicate records, inconsistent formatting, and outliers. Clean and reliable data is essential for accurate analysis and decision-making.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("AB_NYC_2019.csv")

# Display first 5 rows
df.head()
```

```python
# Dataset shape (rows, columns)
df.shape
```

```python
# Column names and data types
df.info()
```

```python
# Statistical overview
df.describe()
```

## Data Integrity and Missing Values Analysis

In this step, the dataset is examined to identify missing values and assess overall data integrity. Handling missing data is crucial to ensure the accuracy, consistency, and reliability of the dataset before further analysis.

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# Percentage of missing values
(df.isnull().sum() / len(df)) * 100
```

```python
# Visual inspection of missing values
df.isnull().any()
```

## Handling Missing Values

Missing values can negatively impact analysis results. In this step, appropriate strategies are applied to handle missing data, such as removing columns with excessive missing values and imputing reasonable values where necessary.

```python
# Check columns with missing values
df.isnull().sum()
```

```python
# Fill missing name
df['name'] = df['name'].fillna("Unknown")
```

```python
# Fill missing reviews_per_month with 0
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
```

```python
# Drop rows with missing last_review
df.dropna(subset=['last_review'], inplace=True)
```

```python
# Verify missing values after cleaning
df.isnull().sum()
```

## Duplicate Record Identification and Removal

Duplicate records can distort analysis results and lead to inaccurate conclusions. In this step, the dataset is checked for duplicate entries and cleaned to ensure data uniqueness.

```python
# Check number of duplicate rows
df.duplicated().sum()
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Verify duplicates after removal
df.duplicated().sum()
```

## Data Standardization and Formatting

To ensure consistency across the dataset, text fields are standardized and column formats are cleaned. Standardization helps avoid inconsistencies that may affect analysis accuracy.

```python
# Standardize neighbourhood_group and room_type text
df['neighbourhood_group'] = df['neighbourhood_group'].str.title()
df['room_type'] = df['room_type'].str.title()
```

```python
# Remove extra spaces from text columns
df['name'] = df['name'].str.strip()
df['host_name'] = df['host_name'].str.strip()
```

```python
# Check unique values after standardization
df['room_type'].unique(), df['neighbourhood_group'].unique()
```

## Outlier Detection and Treatment

Outliers can significantly affect statistical analysis and model performance. In this step, numerical columns are examined to identify extreme values, and appropriate decisions are made to handle them.

```python
# Inspect price distribution
df['price'].describe()
```

```python
# Identify listings with very high prices
df[df['price'] > 1000].shape
```

```python
# Remove extreme price outliers
df = df[df['price'] <= 1000]
```

```python
# Verify price distribution after outlier removal
df['price'].describe()
```

## Final Dataset Validation and Summary

After completing the data cleaning process, the dataset was reviewed to ensure accuracy, consistency, and reliability. Missing values were handled, duplicates were removed, text data was standardized, and outliers were addressed. The cleaned dataset is now suitable for further analysis and modeling.
